```python
import pandas as pd
data = pd.read_csv("balanced_data_resampled_companies.csv")
# test = pd.read_csv("balanced_data_resampled_companies_train.csv")

data
```

```python
import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
test = []
def detect_past_sentence(sentence):

    sent = list(nlp(sentence).sents)[0]
    # test.append(sent)
    # print(list(nlp(sentence).sents))
    return (
        sent.root.tag_ == "VBD" or sent.root.tag_ == "VBN" or
        any(w.dep_ == "aux" and w.tag_ == "VBD" or w.tag_ == "VBN" for w in sent.root.children))
```

```python
tense_list = []
for sentence in data['sentences']:
  tense_list.append(detect_past_sentence(sentence))
  # break

data['past tense'] = tense_list
  
```

BIGRAM/N-GRAM FEATURE

```python
from textblob import TextBlob
import nltk
nltk.download('punkt')
```

```python
irr = pd.read_csv("extracted_irrelevant_sentences.csv")
rel = pd.read_csv("extracted_relevant_sentences.csv")
irr.tail()
```

```python
# get the bigrams collection by tokenizing
bag1 = []
bag2 = []
```

```python
for sentence in irr['relevant_sentences']:
  blobbed1 = TextBlob(sentence)
  bag1.append(blobbed1.ngrams(n=3))

for sentence in rel['relevant_sentences']:
  blobbed2 = TextBlob(sentence)
  bag2.append(blobbed2.ngrams(n=3))

# print(bag)
```

```python
bag1[10]
```

```python
# look at frequencies
def get_freqs(freq, bag):
  for sentence in bag:
    for gram in sentence:
      gram = tuple(gram)
      if freq.get(gram, -1) == -1:
        # count = get_key(gram)
        freq[gram] = 1
      else:
        freq[gram] = freq.get(gram)+1

```

```python
freq_rel = {}
freq_irr = {}

get_freqs(freq_rel, bag2)
get_freqs(freq_irr, bag1)

by_freq_rel = sorted(freq_rel.items(), key=lambda x:x[1])
by_freq_irr = sorted(freq_irr.items(), key=lambda x:x[1])
```

```python
rel_gram = pd.DataFrame(by_freq_rel)
irr_gram = pd.DataFrame(by_freq_irr)
```

```python
rel_gram = rel_gram.sort_values(by = 1, ascending=False)
irr_gram = irr_gram.sort_values(by = 1, ascending=False)
```

```python
rel_gram = rel_gram.rename(columns={0:'n-gram', 1:'count'})
irr_gram = irr_gram.rename(columns={0:'n-gram', 1:'count'})
```

```python
#merge the two dfs and look at count differences
merged = irr_gram.merge(rel_gram, how='outer', on='n-gram')
merged = merged.fillna(0)
```

```python
merged['difference'] = merged['count_x']-merged['count_y']
merged['perc difference'] = abs(merged['difference']/(merged['count_x']+merged['count_y']))
```

```python
merged = merged.rename(columns = {'count_x': 'irr_count', 'count_y': 'rel_count'})
```

```python
merged
```

```python
merged = merged.loc[merged['perc difference'] > 0.3]
merged = merged.loc[merged['irr_count']+merged['rel_count'] > 100.0]
test = merged
```

```python
test['n-gram'] = [str(gram[0]) + ', ' + str(gram[1]) + ', ' + str(gram[2]) for gram in test['n-gram']]
```

```python
test = test.sort_values(by = 'perc difference', ascending=False)
```
