```python
# pull in the csv data

import pandas as pd

relevant = pd.read_csv("../relevant_irrelevant_sentences_labeled/extracted_relevant_sentences.csv")
irrelevant = pd.read_csv("../relevant_irrelevant_sentences_labeled/extracted_irrelevant_sentences.csv")

print(len(irrelevant))
print(len(relevant))
irrelevant["class"] = 0
relevant["class"] = 1


relevant = relevant[relevant.columns[1:]]
irrelevant = irrelevant[irrelevant.columns[1:]]

irrelevant.head()
```

```python
#code for regeneration of dataset after examining misclassified samples

all_sentences = pd.concat([relevant, irrelevant], ignore_index=True)
print(all_sentences.head())

misclassified = pd.read_csv('misclassified_sentences_annotation.csv')
misclassified.columns = misclassified.iloc[0]
misclassified.drop([0], axis=0, inplace=True)

misclassified = misclassified[['Sentences', 'Company', 'True Label', 'Classified by Hand']]

#print(misclassified.head())

label_dict = {'rel': 1, 'irr': 0}

all_sentences_relabeled = pd.DataFrame()

new_relevant = pd.DataFrame()
new_irrelevant = pd.DataFrame()
mis = False

for row in all_sentences.iterrows():
    mis = False
    for misrow in misclassified.iterrows():
        
        if(misrow[1]['Sentences'] == row[1]['sentences'] and misrow[1]['True Label'] != misrow[1]['Classified by Hand']):
            mis = True

            if(misrow[1]['Classified by Hand'] == 'rel'):
                new = row[1]
                new['class'] = 1
                #print(new.to_frame())
                new_relevant = new_relevant.append([new.to_frame().T], ignore_index = True)
                #print(new_relevant)
                break
            if(misrow[1]['Classified by Hand'] == 'irr'):
                new = row[1]
                new['class'] = 0
                #print(new.to_frame())
                new_irrelevant = new_irrelevant.append([new.to_frame().T], ignore_index = True)
                break
            
    if(not mis):
        if(row[1]['class'] == 1 or row[1]['class'] == '1'):
            new_relevant = new_relevant.append([row[1]], ignore_index = True)
        
        if(row[1]['class'] == 0 or row[1]['class'] == '0'):
            new_irrelevant = new_irrelevant.append([row[1]], ignore_index = True)
    
    print(len(new_relevant))
    print(len(new_irrelevant))


print(len(new_relevant))
print(len(new_irrelevant))
```

```python

import random

company_numbers = {}
for row in new_relevant.iterrows():
    if(row[1]['company_label'] not in company_numbers.keys()):
        company_numbers[row[1]['company_label']] = 1
    else:
        company_numbers[row[1]['company_label']] += 1
        
new_relevant_sentences_dict = {}
for row in new_relevant.iterrows():
    l = list(new_relevant_sentences_dict.keys())
    #print(row[1]['relevant_sentences'])
    if row[1]['company_label'] in l:
        new_relevant_sentences_dict[row[1]['company_label']].append(row[1]['sentences'])
    else:
        new_relevant_sentences_dict[row[1]['company_label']] = [row[1]['sentences']]
        

new_irrelevant_sentences_dict = {}
for row in new_irrelevant.iterrows():
    l = list(new_irrelevant_sentences_dict.keys())
    if row[1]['company_label'] not in l:
        new_irrelevant_sentences_dict[row[1]['company_label']] = [row[1]['sentences']]
    else:
        new_irrelevant_sentences_dict[row[1]['company_label']].append(row[1]['sentences'])

        

df_balanced_resampled_new = pd.DataFrame()


for item in new_relevant_sentences_dict.items():
    
    df1 = pd.DataFrame({'sentences': item[1], 
                        'label': [1]*len(item[1]),
                        'company': item[0]
                       })

    df2 = pd.DataFrame({'sentences': random.sample(new_irrelevant_sentences_dict[item[0]], len(item[1])), 
                        'label': [0]*len(item[1]),
                        'company': item[0]
                       })

    df_balanced_resampled_new = pd.concat([df_balanced_resampled_new, df1, df2])
        
print(df_balanced_resampled_new.groupby('label').size())

df_balanced_resampled_new.to_csv("iteration_2/balanced_data_resampled_new.csv")        
```

```python
# sample the irrelevant data so it is balanced with the relevant

sample_irr = irrelevant.sample(n=885, random_state=1)
sample_irr.head()
len(sample_irr)
```

```python
#create the final df
concat = pd.concat([relevant, sample_irr], ignore_index=True)
len(concat)

concat['class']
```

```python
# train and test sets and export to csvs
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(concat['0'], concat['class'], test_size=0.1, random_state=100)
```

```python
# export to csvs

dfbalanced = pd.concat([concat], ignore_index=True)
dfbalanced.to_csv('balanced_data.csv')

dfbalanced_train = pd.concat([X_train.to_frame(), y_train.to_frame()], axis = 1, ignore_index = True)
dfbalanced_train.columns = ['sentences', 'class']
dfbalanced_train.to_csv('balanced_data_train.csv')

dfbalanced_test = pd.concat([X_test.to_frame(), y_test.to_frame()], axis = 1, ignore_index = True)
dfbalanced_test.columns = ['sentences', 'class']
dfbalanced_test.to_csv('balanced_data_test.csv')
```
